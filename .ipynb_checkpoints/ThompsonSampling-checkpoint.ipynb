```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv('Ads_CTR_Optimisation.csv')
veriler
```

```python
import random
```

```python
N = 10000 
d = 10
toplam = 0 
secilenler = []
birler = [0] * d
sifirlar = [0] * d
```

```python
for n in range(1,N):
    ad = 0
    max_th = 0
    for i in range(1,d):
        rasbeta = random.betavariate(birler[i] + 1 , sifirlar[i] + 1)
        if rasbeta > max_th:
            max_th = rasbeta
            ad = i
    secilenler.append(ad)
    odul = veriler.values[n,ad]
    if odul == 1:
        birler[ad] = birler[ad] + 1
    else:
        sifirlar[ad] = sifirlar[ad] + 1
    toplam = toplam + odul
print("Toplam Ödül: ", toplam)
```
