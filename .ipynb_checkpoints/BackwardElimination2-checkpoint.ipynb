```python
#SÃ¼tunlarÄ± 0 1'lere Ã§evirmenin kÄ±sa yolu
from sklearn import preprocessing
veriler2 = veriler.apply(preprocessing.LabelEncoder().fit_transform)
veriler2
```

```python
veriler3 = veriler2.iloc[:, :1]
```

```python
# Ä°lk sÃ¼tunu da 0 1'lere dÃ¶nÃ¼ÅtÃ¼rmÃ¼Å olduk
ohe = preprocessing.OneHotEncoder()
veriler3 = ohe.fit_transform(veriler3).toarray()
veriler3
```

```python
havaDurumu = pd.DataFrame(data = veriler3 ,index = range(14),columns = ["o", "r", "s"])
```

```python
sonDurum = pd.concat([havaDurumu, veriler.iloc[:, 1:3]] , axis =1)
sonDurum
```

```python
sonDurum = pd.concat([veriler2.iloc[:, -2:],sonDurum], axis =1)
sonDurum
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(sonDurum.iloc[:, :-1],sonDurum.iloc[:, -1:], test_size = 0.33 , random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
y_pred
```
