```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv("pca_iris.data",names = ["sepal length" , "sepal width" , "petal length" , "petal width" , "target"])
veriler
```

```python
features =  ["sepal length" , "sepal width" , "petal length" , "petal width" ]
x = veriler[features]
y = veriler[["target"]]
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ["principal component 1", "principal component 2"])
principalDf
```

```python
final_dataframe = pd.concat([principalDf, veriler[["target"]]], axis =1)
final_dataframe
```

```python
dfSetosa = final_dataframe[veriler.target == "Iris-setosa"]
dfVirginica = final_dataframe[veriler.target == "Iris-virginica"]
dfVersicolor = final_dataframe[veriler.target == "Iris-versicolor"]
plt.xlabel = ("principal component 1")
plt.ylabel = ("principal component 2")
plt.scatter(dfSetosa["principal component 1"],dfSetosa["principal component 2"])
plt.scatter(dfVirginica["principal component 1"],dfVirginica["principal component 2"])
plt.scatter(dfVersicolor["principal component 1"],dfVersicolor["principal component 2"])
```

**Loop ile daha kÄ±sa yoldan Ã§izim yapalÄ±m**

```python
targets= ["Iris-setosa","Iris-virginica", "Iris-versicolor"]
colors = ["g", "b", "y"]
plt.xlabel = ("principal component 1")
plt.ylabel = ("principal component 2")

for targ, col in zip(targets,colors):
    dftemp = final_dataframe[veriler.target == targ]
    plt.scatter(dftemp["principal component 1"], dftemp["principal component 2"],color = col)
```

```python
pca.explained_variance_ratio_ # Her bir ana bileÅŸenin veri setindeki toplam varyansÄ±n yÃ¼zdesini temsil eden bir dizi deÄŸeri iÃ§erir.
```
