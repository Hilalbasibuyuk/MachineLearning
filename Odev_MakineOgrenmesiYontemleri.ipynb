```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
```

```python
veriler = pd.read_csv("maaslar_yeni.csv")
veriler
```

```python
x = veriler.iloc[:,2:5].values
y = veriler.iloc[:,5:].values
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
import statsmodels.api as sm
model = sm.OLS(lr.predict(x),x)
model.fit().summary()
#R-squared 1'e yaklaştıkça o kadar güzel bir sonuç almışız demektir
```

```python
r2_score(y, lr.predict(x))
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree =4)
x_poly = pr.fit_transform(x)
x_poly
```

```python
lr2 = LinearRegression()
lr2.fit(x_poly,y)
```

Tahminler

```python
model2 = sm.OLS(lr2.predict(pr.fit_transform(x)),x)
model2.fit().summary()
```

```python
r2_score(y,lr2.predict(pr.fit_transform(x)))
```

Verilerin ölçeklenmesi ile SVR yöntemi

```python
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
x_olcekli = sc1.fit_transform(x)
```

```python
sc2 = StandardScaler()
y_olcekli = sc2.fit_transform(y) # y_olcekli = np.ravel(sc2.fit_transform(Y.reshape(-1,1)))
```

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel = "rbf")
svr_reg.fit(x_olcekli,y_olcekli)
```

```python
model3 = sm.OLS(svr_reg.predict(x_olcekli), x_olcekli)
model3.fit().summary()
```

```python
r2_score(y_olcekli,svr_reg.predict(x_olcekli))
```

Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state = 0)
dtr.fit(x,y)
```

```python
model4 = sm.OLS(dtr.predict(x),x)
model4.fit().summary()
```

```python
r2_score(y,dtr.predict(x))
```

Random Forest Regresyonu

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators = 10 ,random_state = 0)
rf_reg.fit(x,y.ravel())
```

```python
model5 = sm.OLS(rf_reg.predict(x),x)
model5.fit().summary()
```

```python
r2_score(y,rf_reg.predict(x))
```
