```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv("illness.csv")
veriler
```

```python
Gender = veriler.iloc[:,0:1].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
Gender[:,0] = le.fit_transform(veriler.iloc[:,0])
Gender
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
Gender = ohe.fit_transform(Gender).toarray()
Gender
```

```python
sonuc = pd.DataFrame(data = Gender[:,:1], index = range(54), columns = ["Gender"])
sonuc
```

```python
veriler["Gender"] = sonuc
veriler
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(veriler[["RR"]], veriler["Condition"])
```

```python
lr.predict([[20]])
```

```python
lr2 = LinearRegression()
lr2.fit(veriler[["Gender", "RR", "SPO2"]], veriler["Condition"])
```

```python
lr2.predict([[1,18,95]])
```
