```python
import numpy as np
import pandas as pd
```

```python
column_names = ["user_id", "item_id", "rating", "timestamp"]
df = pd.read_csv("users.data", sep = "\t" , names = column_names)
df
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles
```

```python
df = pd.merge(df,movie_titles, on = "item_id")
df
```

```python
moviemat = df.pivot_table(index = "user_id", columns = "title", values = "rating")
moviemat
# User_id'yi index olarak kullandık. Title'ları direkt columns yaptık. Tablodaki değerler ise rating oranları.
```

### Amaç starwars filmine benzer film önerileri yapmak

```python
starwars_user_ratings = moviemat["Star Wars (1977)"]
starwars_user_ratings
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_starwars
```

```python
type(similar_to_starwars)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ["Correlation"])
corr_starwars.dropna(inplace = True)
corr_starwars # Korelasyondaki nanları sildik ve serilerden dataframe'e dönüştürdük
```

```python
corr_starwars.sort_values("Correlation", ascending = False).head(10)
```

Yukarıda alakasız sonuçlar çıktı. Çünkü Bazı filmler çok az oy almış aslında ve eşdeğer değiller. Bu yüzden 100'den az oy alan filmleri eleyelim.

```python
df.drop(["timestamp"],axis = 1) # Gereksiz diye sildik
```

```python
ratings = pd.DataFrame(df.groupby("title") ["rating"].mean())
ratings
```

```python
ratings.sort_values("rating", ascending = False).head()
```

**Filmlerin aldığı oy sayısını bulalım**

```python
ratings["rating_oy_sayisi"] = pd.DataFrame(df.groupby("title")["rating"].count())
ratings.head(10)
```

```python
ratings.sort_values("rating_oy_sayisi",ascending = False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings["rating_oy_sayisi"])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars["rating_oy_sayisi"]>100].sort_values("Correlation", ascending = False).head()
```

**Ve sonuç olarak Star Wars filmine benzer filmleri sıralamış olduk**
